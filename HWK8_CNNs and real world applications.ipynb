```python
# Descarga del dataset
import kagglehub

# Download latest version
path = kagglehub.dataset_download("ashwingupta3012/male-and-female-faces-dataset")

print("Path to dataset files:", path)
```

```python
# Configuración de parámetros
IMG_SIZE = (200, 200)  # Las imágenes originales son 200x200
BATCH_SIZE = 32
EPOCHS = 50
```

```python
# Cargar y preparar los datos
def load_dataset(base_path):
    # Primero inspeccionamos la estructura del directorio
    print("Contenido del directorio base:", os.listdir(base_path))

    # Buscamos las carpetas que contienen las imágenes
    for root, dirs, files in os.walk(base_path):
        print(f"Directorio actual: {root}")
        print(f"Subdirectorios: {dirs}")
        print(f"Archivos: {files[:5]}...")  # Mostrar solo primeros 5 archivos

        # Buscamos directorios que puedan contener las imágenes
        if 'Male Faces' in dirs or 'Female Faces' in dirs:
            male_path = os.path.join(root, 'Male Faces')
            female_path = os.path.join(root, 'Female Faces')
            break
        elif 'male' in dirs or 'female' in dirs:
            male_path = os.path.join(root, 'Male Faces')
            female_path = os.path.join(root, 'Female Faces')
            break
        elif 'Male Faces' in dirs or 'Female Faces' in dirs:
            male_path = os.path.join(root, 'Male Faces')
            female_path = os.path.join(root, 'Female Faces')
            break
    # Verificamos si encontramos los paths  # This line was indented, causing the error.
    if not os.path.exists(male_path) or not os.path.exists(female_path):
        raise FileNotFoundError(f"No se encontraron los directorios 'Male' y 'Female' en {base_path}")

    images = []
    labels = []
```

```python
    # Cargar imágenes male (etiqueta 0)
    male_files = [f for f in os.listdir(male_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    for img_file in male_files[:2000]:  # Limitar para equilibrio
        img_path = os.path.join(male_path, img_file)
        try:
            img = load_img(img_path, target_size=IMG_SIZE)
            img_array = img_to_array(img) / 255.0
            images.append(img_array)
            labels.append(0)
        except Exception as e:
            print(f"Error cargando {img_path}: {e}")

    # Cargar imágenes female (etiqueta 1)
    female_files = [f for f in os.listdir(female_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    for img_file in female_files[:2000]:  # Limitar para equilibrio
        img_path = os.path.join(female_path, img_file)
        try:
            img = load_img(img_path, target_size=IMG_SIZE)
            img_array = img_to_array(img) / 255.0
            images.append(img_array)
            labels.append(1)
        except Exception as e:
            print(f"Error cargando {img_path}: {e}")

    return np.array(images), np.array(labels)
```
